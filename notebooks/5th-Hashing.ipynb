```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
N = 2**18
learning_rate = 0.02
weights = [0]*N
```

```python
def hash(a):
    offset = 0x811c9dc5
    fnv_32_prime = 0x01000193
    size = 2 ** 32
    for b in a:
        offset = offset ^ ord(b)
        offset = (offset * fnv_32_prime) % size
    return offset
```

```python
def hashing(features , N):
    x = []
    for index , f in features.items():
        h = hash(f)
        x.append(h%N)
    return x
```

```python
def prediction(x , weights):
    w_x_product=0
    for i in x:
        w_x_product+=weights[i]
    return 1/(1+np.exp(-w_x_product))
```

```python
def logloss(p , y , eps=1e-15):
    pred = np.clip(p,eps,1-eps)
    if y==1:
        return -np.log(pred)
    else:
        return -np.log(1-pred)
```

```python
train_file = csv.DictReader(open('../datadays2020/click_train.csv'))
counter = 0
current_progress = 0
for features in train_file:
    counter+=1
    del(features['adId'])
    y = float(features['clicked'])
    del(features['clicked'])
    x = hashing(features , N)
    
    for i in x:
        weights[i] = -(prediction(x,weights)-y)*learning_rate
    current_progress += logloss(prediction(x,weights) , y)
    if counter % 10000 == 0:
        print("training , loss: " , round(current_progress/counter,4))
```
