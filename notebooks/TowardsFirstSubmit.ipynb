## probability approach


```python
ad_prob={}
#user_count={}
for i , r in df.iterrows():
    if r['adId'] in ad_prob:
        ad_prob[r['adId']] += r['clicked']
    else:
        ad_prob[r['adId']] = r['clicked']
```

```python
dftest = pd.read_csv('../datadays2020/click_test.csv')
```

```python
user_count={}
for i , r in df.iterrows():
    if r['userId'] in user_count:
        user_count[r['userId']] += r['clicked']
    else:
        user_count[r['userId']] = r['clicked']
```

```python
diad={}
for i,r in dftest.iterrows():
    if r['displayId'] in diad:
        diad[r['displayId']].append(r['adId'])
    else:
        diad[r['displayId']] = [r['adId'],]
```

```python
for i in diad:
    diad[i] = sorted(diad[i] , key =lambda x: -1 * ad_prob.get(x) if x in ad_prob else 0)
```

```python
dftest['ranked'] = dftest.apply(lambda x : diad[x['displayId']].index(x['adId']) + 1 , axis = 1)
```
