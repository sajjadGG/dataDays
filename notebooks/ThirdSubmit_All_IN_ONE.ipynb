```python
def create_real_train(features,column_names,x_train , word_prob=None , ignore=['timestamp']):
    if word_prob!=None:
        features.remove('words')
    count=0;
    n = len(x_train)
    print(n)
    dfdummy = pd.DataFrame(columns = column_names)
    featuregroups = {}
    for f in features:
        featuregroups[f] = dftrain.groupby(f).mean()['clicked']
    temps=[]
    for i,r in x_train.iterrows():
        count+=1
        if(count%100000==0):
            print("--{}--".format(count/n))
        temp=[]
        for f in features:
            if f in ignore:
                temp.append(r[f])
                continue
            tr = featuregroups[f].get(r[f])
            #print(tr)
            #print('---------{}-----'.format(r[f]))
            temp.append(tr if tr!=None else 0)
        temps.append(temp)
    dfdummy=dfdummy.append(pd.DataFrame(temps, columns=column_names) , ignore_index=True)
    return dfdummy
```

```python
def create_real_train2(features,column_names,x_train ,  ignore=['timestamp']):
    count=0;
    n = len(x_train)
    print(n)
    dfdummy = pd.DataFrame(columns = column_names)
    featuregroups = {}
    for f in features:
        featuregroups[f] = dftrain.groupby(f).mean()['clicked']
        
    for i in range(len(features)) :
        f1 = features[i]
        if f1 in ignore:
            continue
        for j in range(i+1 , len(features)):
            f2 = features[j]
            if f2 in ignore:
                continue
            featuregroups[(f1,f2)] = dftrain.groupby([f1,f2]).mean()['clicked']
            
    temps=[]
    #print(featuregroups.keys())
    for i,r in x_train.iterrows():
        count+=1
        if(count%100000==0):
            print("--{}--".format(count/n))
        temp=[]
        for f in features:
            if f in ignore:
                temp.append(r[f])
                continue
            tr = featuregroups[f].get(r[f])
            temp.append(tr if tr!=None else 0)
        for j in range(len(features)) :
            f1 = features[j]
            if f1 in ignore:
                continue
            for k in range(j+1 , len(features)):
                f2 = features[k]
                if f2 in ignore:
                    continue
                if f1==f2:
                    continue
                tr = featuregroups[(f1,f2)].get((r[f1],r[f2]))
                #print(tr)
                temp.append(tr if tr!=None else 0)
                
        temps.append(temp)
    dfdummy=dfdummy.append(pd.DataFrame(temps) , ignore_index=True)
    return dfdummy
```

```python
def create_real_train3(anchor , features,column_names,x_train , word_prob=None , ignore=['timestamp']):
    if word_prob!=None:
        features.remove('words')
    count=0;
    n = len(x_train)
    print(n)
    dfdummy = pd.DataFrame(columns = column_names)
    featuregroups = {}
    featuregroups[anchor] = dftrain.groupby(anchor).mean()['clicked']
    for f in features:
        featuregroups[f] = dftrain.groupby(f).mean()['clicked']
    temps=[]
    for i,r in x_train.iterrows():
        count+=1
        if(count%100000==0):
            print("--{}--".format(count/n))
        tr = featuregroups[anchor].get(r[anchor])
        temp=[tr if tr!=None else 0]
        for f in features:
            if f in ignore:
                temp.append(r[f])
                continue
            tr = featuregroups[f].get((r[f[0]] , r[f[1]]))
            #print(tr)
            #print('---------{}-----'.format(r[f]))
            temp.append(tr if tr!=None else 0)
        temps.append(temp)
    dfdummy=dfdummy.append(pd.DataFrame(temps, columns=column_names) , ignore_index=True)
    return dfdummy
```

```python
def create_real_train4(anchors , features,column_names,x_train ):
    count=0;
    n = len(x_train)
    print(n)
    dfdummy = pd.DataFrame(columns = column_names)
    featuregroups = {}
    for anchor in anchors:
        featuregroups[anchor] = dftrain.groupby(anchor).mean()['clicked']
    for f in features:
        featuregroups[f] = dftrain.groupby(f).mean()['clicked']
    temps=[]
    for i,r in x_train.iterrows():
        count+=1
        if(count%100000==0):
            print("--{}--".format(count/n))
        temp=[]
        for anchor in anchors:
            tr = featuregroups[anchor].get(r[anchor])
            temp.append(tr if tr!=None else 0)
        for f in features:
            tr = featuregroups[f].get((r[f[0]] , r[f[1]]))
            temp.append(tr if tr!=None else 0)
        temps.append(temp)
    dfdummy=dfdummy.append(pd.DataFrame(temps, columns=column_names) , ignore_index=True)
    return dfdummy
```

```python
def dropout(X , alpha):
    n = int(alpha * len(X))
    l = np.random.randint(len(X) , size = n)
    l2 = np.random.randint(X.shape[1] , size=n)
    for i in range(n):
        X.loc[l[i]][X.columns[l2[i]]] = 0
```
